## Preparação dos dados

```python
train_path = 'dataset/train'
valid_path = 'dataset/valid'
test_path = 'dataset/test'
```

```python
## Convertendo as imagens para um formato adequado que o modelo Sequential espera para fazer o treinamento. Formato de keras generator vgg16
train_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input).flow_from_directory(directory=train_path, target_size=(224,224), classes=['azalea', 'camellia', 'sunflower'], batch_size=10)
valid_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input).flow_from_directory(directory=valid_path, target_size=(224,224), classes=['azalea', 'camellia', 'sunflower'], batch_size=10)
test_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input).flow_from_directory(directory=test_path, target_size=(224,224), classes=['azalea', 'camellia', 'sunflower'], batch_size=10, shuffle=False) ## É importante não é para embaralhar pq será útil na matriz de confusão
```

```python
imgs, labels = next(train_batches) ## Pegando um único batch de 10 imagens e labels do dataset de treinamento
```

```python
## Funcão para plotar as imagens em grid com 1 linha e 10 colunas
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
plotImages(imgs)
print(labels)
## As imagens vem com cores distorcidas devido ao preprocessamento para mudar seu formato
```

## Treinamento do modelo

```python
model = Sequential([
    Conv2D(filters=32, kernel_size=(3, 3), activation='relu', padding = 'same', input_shape=(224,224,3)), ## padding='same' a dimensão da imagem não reduz. input_shape é a primeira camada de entrada de dados
    MaxPool2D(pool_size=(2, 2), strides=2), ## maxpooling vai reduzir o tamanho da imagem por 2
    Conv2D(filters=64, kernel_size=(3, 3), activation='relu', padding = 'same'),
    MaxPool2D(pool_size=(2, 2), strides=2),
    Flatten(), ## transformando para um tensor de uma única dimensão
    Dense(units=3, activation='softmax') ## dois nós/neurônios correspondendo cats and dogs. A função de ativição Softmax dará as probabilidades de cada classe
])
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x=train_batches, ## Nesse caso não precisou passar as labels Y, pois elas já se encontram dentro de X
    steps_per_epoch=len(train_batches),
    validation_data=valid_batches,
    validation_steps=len(valid_batches),
    epochs=10,
    verbose=2
)
```

## Previsão

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.title('Loss')
plt.xlabel('Epoch')
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['accuracy', 'val_accuracy'])
plt.title('Accuracy')
plt.xlabel('Epoch')
```

```python
## Função extraída dos docs scikitlearn
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
test_batches.class_indices
```

```python
cm_plot_labels = ['azalea','camellia','sunflower']
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title='Confusion Matrix')
```

## Utilizando um modelo pré-treinado

```python
vgg16_model = tf.keras.applications.vgg16.VGG16()
```

```python
vgg16_model.summary()
```

```python
type(vgg16_model)
```

```python
model = Sequential()
for layer in vgg16_model.layers[:-1]:
    model.add(layer)
```

```python
for layer in model.layers:
    layer.trainable = False
```

```python
model.add(Dense(units=3, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=train_batches,
          steps_per_epoch=len(train_batches),
          validation_data=valid_batches,
          validation_steps=len(valid_batches),
          epochs=5,
          verbose=2
)
```

```python
predictions = model.predict(x=test_batches, steps=len(test_batches), verbose=0)
```

```python
cm = confusion_matrix(y_true=test_batches.classes, y_pred=np.argmax(predictions, axis=-1))
cm_plot_labels = ['azalea','camellia','sunflower']
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title='Confusion Matrix')
```
